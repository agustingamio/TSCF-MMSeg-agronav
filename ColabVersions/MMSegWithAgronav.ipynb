```python
import os

def clean_unpaired_files(image_dir, label_dir):
    image_files = {os.path.splitext(f)[0] for f in os.listdir(image_dir) if f.endswith('.JPG')}
    label_files = {os.path.splitext(f)[0] for f in os.listdir(label_dir) if f.endswith('.png')}

    # Find unpaired files
    images_only = image_files - label_files
    labels_only = label_files - image_files

    # Delete unmatched images
    for fname in images_only:
        path = os.path.join(image_dir, f'{fname}.JPG')
        if os.path.exists(path):
            os.remove(path)
            print(f"Deleted unmatched image: {path}")

    # Delete unmatched labels
    for fname in labels_only:
        path = os.path.join(label_dir, f'{fname}.png')
        if os.path.exists(path):
            os.remove(path)
            print(f"Deleted unmatched label: {path}")

    print("✅ Cleanup complete.")

if __name__ == "__main__":
    image_dir = '/content/TSCF-MMSeg-agronav/data/images'
    label_dir = '/content/TSCF-MMSeg-agronav/data/labels'

    clean_unpaired_files(image_dir, label_dir)
```

```python
from mmseg.apis import init_model, inference_model
import mmcv
import numpy as np
import matplotlib.pyplot as plt

# Config and checkpoint
config_file = '/content/TSCF-MMSeg-agronav/configs/agronav/agronav-deeplabv3plus-resnet50.py'
checkpoint_file = '/content/TSCF-MMSeg-agronav/work_dirs/agronav-deeplabv3plus-resnet50/iter_16000.pth'

# Init model
model = init_model(config_file, checkpoint_file, device='cuda:0')

# Load image
img_path = '/content/c3_ZoLPlT.jpg'
img = mmcv.imread(img_path, channel_order='rgb')

# Run inference
result = inference_model(model, img)

# Extract segmentation map from PixelData object
# Access the data field of the PixelData object and convert to numpy
seg_map = result.pred_sem_seg.data.cpu().numpy()
print(f"Seg_map shape: {seg_map.shape}, dtype: {seg_map.dtype}")

# Visualize manually
img = mmcv.imread(img_path)
img = mmcv.imconvert(img, 'bgr', 'rgb')

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
# Squeeze to remove any singleton dimensions if needed
seg_map_vis = seg_map.squeeze()
plt.imshow(seg_map_vis, cmap='viridis')  # Using a colormap for better visualization
plt.title('Segmentation Map')
plt.axis('off')

plt.tight_layout()
plt.show()
```
